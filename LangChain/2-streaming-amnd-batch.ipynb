```python
import os
from langchain.chat_models import init_chat_model
```

```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

llm = ChatNVIDIA(model="openai/gpt-oss-20b")
```

```python
for chunk in llm.stream("Write me a story with max 500 words"):
    print(chunk.text, end = "|", flush = True)
```

```python
responses = llm.batch([
    "Why do parrots have a colorful feathers?",
    'How do airplanes fly?',
    "Where do zebras live?"
])

for response in responses:
    print(response)
```

```python
responses = llm.batch([
    "Why do parrots have a colorful feathers?",
    'How do airplanes fly?',
    "Where do zebras live?",
    "Where do you live?",
    "Who will win the World Cup 2026 in your opinion??",

],
config={
    'max_concurrency' : 4
}
)

for response in responses:
    print(response)
```
