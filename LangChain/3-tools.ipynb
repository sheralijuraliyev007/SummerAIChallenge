```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

llm = ChatNVIDIA(model="openai/gpt-oss-20b")
```

```python
from langchain.tools import tool
```

```python
@tool
def get_weather(location:str) -> str:
    """Get weather at a location"""
    return f"It is sunny in {location}"
```

```python
model_with_tools = llm.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("What is the weather like in Boston")
print(response)
for tool_call in response.tool_calls:
    print(f"Tool: {tool_call['name']}")
    print(f"Args: {tool_call['args']}")
```

### Tool Execution loops

```python
messages = [{"role":"user","content":"What is the weather in Boston?"}]
ai_messages = model_with_tools.invoke(messages)
#print(ai_messages)
messages.append(ai_messages)


for tool_call in ai_messages.tool_calls:
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)

final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
